# 🔹 2.1 DAGs Page (🏠 Home)

## 📌 What it shows:
- List of all DAGs
- DAG status (on/off)
- Last run status
- Next scheduled run

---

## 🧠 Features:

### ✅ Toggle DAG
- Enable/disable DAG execution

---

### ✅ Run DAG Manually
- Trigger DAG instantly

---

### ✅ DAG Filters
- Filter by:
  - Status
  - Tags
  - Owners

---

### ✅ DAG Health Indicators
- Green → success
- Red → failed
- Yellow → running
