# 🔥 3. Classic DAG Example

```python
from airflow import DAG
from airflow.operators.python import PythonOperator
from airflow.utils.task_group import TaskGroup
from datetime import datetime, timedelta

default_args = {
    "owner": "airflow",
    "retries": 2,
    "retry_delay": timedelta(minutes=5)
}

def extract(ti):
    data = "sample data"
    ti.xcom_push(key="data", value=data)

def transform(ti):
    data = ti.xcom_pull(task_ids="extract", key="data")
    print(f"Transforming {data}")

def load():
    print("Loading data")

with DAG(
    dag_id="classic_dag",
    start_date=datetime(2024, 1, 1),
    schedule_interval="@daily",
    catchup=False,
    default_args=default_args,
    max_active_tasks=3
) as dag:

    with TaskGroup("group") as group:
        t1 = PythonOperator(task_id="extract", python_callable=extract)
        t2 = PythonOperator(task_id="transform", python_callable=transform)

    t3 = PythonOperator(task_id="load", python_callable=load)

    group >> t3
```

# ⚡ 4. Decorator DAG Example

```python
from airflow.decorators import dag, task
from datetime import datetime, timedelta

@dag(
    dag_id="decorator_dag",
    start_date=datetime(2024, 1, 1),
    schedule_interval="@daily",
    catchup=False,
    default_args={
        "retries": 2,
        "retry_delay": timedelta(minutes=5)
    }
)
def etl_pipeline():

    @task
    def extract():
        return "sample data"

    @task
    def transform(data):
        return data.upper()

    @task
    def load(data):
        print(f"Loading {data}")

    data = extract()
    transformed = transform(data)
    load(transformed)

dag = etl_pipeline()
```

# 🔹 3. PythonOperator

## Classic

```python
from airflow import DAG
from airflow.operators.python import PythonOperator
from datetime import datetime

def my_func():
    print("Hello")

with DAG("python_classic", start_date=datetime(2024,1,1), schedule="@daily", catchup=False) as dag:
    task = PythonOperator(task_id="task", python_callable=my_func)
```

## Decorator

```python
from airflow.decorators import dag, task
from datetime import datetime

@dag(start_date=datetime(2024,1,1), schedule="@daily", catchup=False)
def dag_example():

    @task
    def my_func():
        print("Hello")

    my_func()

dag = dag_example()
```

# 🏗️ 1. Enterprise DAG Example (S3 → Spark → Warehouse)

```python
from airflow.decorators import dag, task
from datetime import datetime

@dag(start_date=datetime(2024,1,1), schedule="@daily", catchup=False)
def enterprise_pipeline():

    @task
    def extract():
        print("Reading from S3")
        return "raw_data"

    @task
    def transform(data):
        print("Processing with Spark")
        return data.upper()

    @task
    def load(data):
        print("Loading to Warehouse")

    load(transform(extract()))

dag = enterprise_pipeline()
```

# 🚀 Airflow Docker Bind Mounts — Explained

---

# 🧠 What is Happening?

Developer → Local DAG Folder → Bind Mount → Docker Container → Airflow DAGs Folder

---

# 🧩 Explanation

## Developer
Creates DAG files locally.

## Local Folder
Example:
./dags

## Bind Mount
Maps local folder to container:

./dags → /opt/airflow/dags

## Docker Container
Airflow reads DAGs from:
/opt/airflow/dags

---

# 🔄 Flow

1. Create DAG locally
2. Saved in ./dags
3. Automatically available in container
4. Airflow loads DAG

  - ./dags:/opt/airflow/dags

# 🧠 Benefits

# 🚀 Airflow Logs — Complete Notes

---

# 🧠 Overview

Airflow generates logs at different stages:

- DAG Parsing
- DAG Run
- Task Execution

---

# 📂 Log Structure

logs/
│

├── dag_processor/

├── dag_id=<dag_name>/

│   ├── run_id=<run_id>/

│   │   ├── task_id=<task_name>/

│   │   │   └── attempt=1.log

---

# 🔹 DAG Processor Logs

Folder:
logs/dag_processor/

Generated when:
- DAG is parsed
- Scheduler reads DAG

Contains:
- Syntax errors
- Import issues

---

# 🔹 DAG Run Logs

Folder:
logs/dag_id=<dag_name>/run_id=<run_id>/

Generated when:
- DAG is triggered

Contains:
- Run metadata

---

# 🔹 Task Logs

Folder:
logs/dag_id=<dag_name>/run_id=<run_id>/task_id=<task_name>/

Generated when:
- Task executes

Contains:
- Print output
- Errors
- Execution details

---

# 🔄 Flow

1. DAG parsed → dag_processor logs
2. DAG triggered → run_id created
3. Task runs → task logs created

---

# ⚙️ Key Concepts

- run_id → unique DAG run
- task_id → specific task
- attempt → retry count

---

# ⚠️ Issues

- Logs missing → worker not running
- Permission issues → fix chmod

---

# 🎯 Summary

Airflow logs are structured by DAG → Run → Task and help in debugging and monitoring.

# 🧠 1. What is Airflow Metadata DB?

The Airflow metadata DB (commonly PostgreSQL or MySQL) stores ALL state of Airflow.

It is the single source of truth for:
- DAG runs
- Task states
- Scheduling decisions
- Logs metadata

---

# 🎯 2. What Data is Stored?

## Core Data
- DAG definitions (serialized)
- DAG runs
- Task instances
- Execution states
- Start/end times

## Operational Data
- Scheduler jobs
- Worker jobs
- Logs references

## Config Data
- Connections
- Variables
- Pools
- Users

psql -h <host> -U <user> -d airflow

# 🧪 3. Production Queries (VERY USEFUL 🔥)

---

## 🔹 DAG Success Rate

```sql
SELECT dag_id,
       COUNT(*) AS total_runs,
       SUM(CASE WHEN state='success' THEN 1 ELSE 0 END) AS success_runs
FROM dag_run
GROUP BY dag_id;
```

---

## 🔹 Failed Tasks

```sql
SELECT dag_id, task_id, execution_date
FROM task_instance
WHERE state='failed';
```

---

## 🔹 Long Running Tasks

```sql
SELECT dag_id, task_id,
       EXTRACT(EPOCH FROM (end_date - start_date)) AS duration
FROM task_instance
ORDER BY duration DESC;
```

---

## 🔹 Currently Running Tasks

```sql
SELECT * FROM task_instance WHERE state='running';
```

---

## 🔹 Recent DAG Runs

```sql
SELECT dag_id, run_id, state, execution_date
FROM dag_run
ORDER BY execution_date DESC
LIMIT 10;
```

# ⚙️ 4. How Airflow Updates These Tables (Internals)

---

## 🔹 Step 1: DAG Parsing

- Scheduler parses DAG
- Updates:
  - serialized_dag
  - dag table

---

## 🔹 Step 2: DAG Run Creation

- Scheduler inserts into:
  - dag_run

---

## 🔹 Step 3: Task Scheduling

- Scheduler creates entries in:
  - task_instance (state = scheduled)

---

## 🔹 Step 4: Task Execution

- Executor sends task to worker
- Worker updates:
  - task_instance.state → running → success/failed

---

## 🔹 Step 5: Logging

- Logs stored in:
  - log table (metadata)
  - filesystem / S3 (actual logs)

---

## 🔹 Step 6: XCom Updates

- Task pushes data → stored in:
  - xcom table
