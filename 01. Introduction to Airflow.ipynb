# 🚀 Apache Airflow

---

# 🧠 1. What is Apache Airflow?

Apache Airflow is an open-source platform used to **author, schedule, and monitor workflows programmatically**.

- Workflows are defined as **DAGs (Directed Acyclic Graphs)**
- Each node = task
- Edges = dependencies

👉 Airflow = Workflow Orchestration Tool

---

# ❗ 2. Problems Before Airflow

## 2.1 Cron Job Chaos
- Scripts scheduled manually
- No dependency handling

## 2.2 No Dependency Management
- Tasks run even if previous fails

## 2.3 No Monitoring
- No UI
- Hard debugging

## 2.4 No Retry Mechanism
- Manual reruns required

## 2.5 Poor Scalability
- Hard to manage large pipelines

## 2.6 No Version Control
- Pipelines not treated as code

---

# 💡 3. How Airflow Solves These Problems

## 3.1 Workflow as Code
- Python-based DAGs
- Version control friendly

## 3.2 DAG-based Execution
- Clear dependencies

## 3.3 Scheduler
- Automates execution

## 3.4 Retry Mechanism
- Automatic retries

## 3.5 Monitoring UI
- Centralized logs and tracking

## 3.6 Scalability
- Multiple executors

---

# 🧩 3. Core Components

---

# 🔹 3.1 DAG (Directed Acyclic Graph)

## 📌 Definition
A DAG defines:
- Tasks
- Dependencies
- Execution order

---

## ⚙️ Characteristics

- Directed → order defined
- Acyclic → no loops allowed
- Graph → nodes + edges

---

## 🧠 Example

task_a >> task_b >> task_c

---

## 🎯 Responsibilities

- Define workflow logic
- Set scheduling rules
- Configure retries

---

# 🔹 3.2 DAG Folder

## 📌 What is it?
- Directory where DAG Python files are stored

---

## ⚙️ Behavior

- Scheduler continuously scans this folder
- Parses DAG files

---

## ⚠️ Important

- Only Python scripts
- Must be lightweight (no heavy computation)

---

# 🔹 3.3 Scheduler (Brain of Airflow)

## 📌 Definition
The Scheduler is responsible for:
- Monitoring DAGs
- Scheduling tasks
- Triggering execution

---

## ⚙️ Internal Working

1. Reads DAG files
2. Creates DAG runs
3. Identifies runnable tasks
4. Sends tasks to executor

---

## 🧠 Key Functions

- Dependency resolution
- Time-based scheduling
- Task state management

---

## ⚠️ Important

Scheduler does NOT execute tasks
It only decides what should run

---

# 🔹 3.4 Executor (Execution Engine)

## 📌 Definition
Responsible for:
- Running tasks
- Allocating resources

---

## ⚙️ Types

- SequentialExecutor
- LocalExecutor
- CeleryExecutor
- KubernetesExecutor

---

## 🧠 Role

- Receives tasks from scheduler
- Sends tasks to workers

---

# 🔹 3.5 Workers

## 📌 Definition
Workers are processes/machines that:
- Execute tasks

---

## ⚙️ Behavior

- Pull tasks from executor
- Run task logic
- Update status

---

## 🎯 Key Role

Actual computation happens here

---

# 🔹 3.6 Metadata Database (Backbone)

## 📌 Definition
Stores all Airflow state

---

## 🧠 Stores:

- DAG runs
- Task states
- Logs (metadata)
- Users & connections

---

## ⚙️ Common Databases

- PostgreSQL
- MySQL

---

## ⚠️ Critical Component

If DB fails → Airflow fails

---

# 🔹 3.7 Web Server (UI Layer)

## 📌 Definition
Provides UI for monitoring

---

## 🧠 Features

- View DAGs
- Trigger DAGs
- Monitor task status
- View logs

---

## 🎯 Importance

- Observability
- Debugging

---

# 🔹 3.8 Message Broker (Celery Only)

## 📌 Definition
Queue system for task distribution

---

## 🧠 Examples

- Redis
- RabbitMQ

---

## ⚙️ Role

- Stores tasks in queue
- Workers pull tasks

---

# 🔹 3.9 Triggerer (Async Tasks)

## 📌 Definition
Handles async/sensor-based tasks

---

## 🧠 Example

- Waiting for file arrival
- API polling

---

## ⚙️ Benefit

- Frees worker resources
- Efficient async execution

# 🚀 Apache Airflow Executors — Complete Detailed Guide

---

# 🧠 1. What is an Executor?

In Apache Airflow, an **Executor** is the component responsible for:

- Running tasks
- Managing execution resources
- Enabling parallelism
- Distributing workload

---

## 🔁 Execution Flow

DAG → Scheduler → Executor → Worker → Task Execution

- **Scheduler** → decides *what* to run
- **Executor** → decides *how & where* to run

---

# 🎯 2. Why Executors Are Critical

Without executors:
- No parallel execution ❌
- No scaling ❌
- No distributed processing ❌

Executors solve:
- Task parallelism
- Resource allocation
- Distributed execution
- Fault tolerance

---

# 🧩 3. Types of Executors

---

# 🔹 3.1 SequentialExecutor

## 📌 Definition
The **SequentialExecutor** is the simplest executor that runs **one task at a time**.

---

## ⚙️ How It Works

- Tasks are executed sequentially
- No parallelism
- Runs inside the scheduler process

Task A → Task B → Task C

---

## ✅ Advantages

- Very simple setup
- No external dependencies
- Good for debugging

---

## ❌ Disadvantages

- Extremely slow
- No scalability
- Not suitable for production

---

## 🎯 Use Cases

- Learning Airflow
- Local development
- Debugging DAGs

---

# 🔹 3.2 LocalExecutor

## 📌 Definition
Executes tasks **in parallel on a single machine** using multiprocessing.

---

## ⚙️ How It Works

- Scheduler spawns multiple worker processes
- Each process runs a task

CPU Core 1 → Task A
CPU Core 2 → Task B
CPU Core 3 → Task C

---

## 🔧 Internal Behavior

- Uses Python multiprocessing
- Controlled by:
  - parallelism
  - worker slots

---

## ✅ Advantages

- Easy to set up
- Parallel execution
- No external systems needed

---

## ❌ Disadvantages

- Limited by machine resources
- No horizontal scaling
- Can overload system

---

## 🎯 Use Cases

- Small to medium workloads
- Single-node production setups

---

# 🔹 3.3 CeleryExecutor

## 📌 Definition
A **distributed executor** that uses Celery to execute tasks across multiple workers.

---

## ⚙️ Architecture

Scheduler → Message Broker → Worker Nodes → Task Execution

---

## 🔧 Components

### 1. Message Broker
- Redis / RabbitMQ
- Stores task queue

### 2. Workers
- Fetch tasks from queue
- Execute tasks

### 3. Result Backend
- Stores task results

---

## ⚙️ How It Works

1. Scheduler pushes task to broker
2. Worker pulls task
3. Worker executes task
4. Status updated in DB

---

## ✅ Advantages

- Horizontal scalability
- Distributed processing
- Fault tolerance
- High availability

---

## ❌ Disadvantages

- Complex setup
- Requires infrastructure:
  - Redis / RabbitMQ
- Debugging can be harder

---

## 🎯 Use Cases

- Large-scale pipelines
- Production environments
- Multi-node systems

---

# 🔹 3.4 KubernetesExecutor

## 📌 Definition
Runs each task in a **separate Kubernetes pod**.

---

## ⚙️ Architecture

Scheduler → Kubernetes API → Pod → Task Execution

---

## ⚙️ How It Works

1. Scheduler requests pod creation
2. Kubernetes creates a pod
3. Task runs inside pod
4. Pod is destroyed after completion

---

## 🔧 Key Features

- One task = one pod
- Dynamic resource allocation
- Isolation per task

---

## ✅ Advantages

- True isolation
- Auto-scaling
- Efficient resource usage
- Cloud-native

---

## ❌ Disadvantages

- Requires Kubernetes knowledge
- Higher setup complexity
- Slight startup latency (pod spin-up)

---

## 🎯 Use Cases

- Cloud-native systems
- Microservices architecture
- Variable workloads

---

# 🔹 3.5 CeleryKubernetesExecutor

## 📌 Definition
A **hybrid executor** combining Celery and Kubernetes.

---

## ⚙️ How It Works

- Some tasks → Celery workers
- Some tasks → Kubernetes pods

---

## ✅ Advantages

- Flexibility
- Cost optimization
- Control over execution type

---

## ❌ Disadvantages

- Very complex setup
- Harder to maintain

---

## 🎯 Use Cases

- Mixed workloads
- Gradual migration to Kubernetes

---

# 🔹 3.6 DebugExecutor

## 📌 Definition
Executor used for **debugging DAGs**.

---

## ⚙️ Behavior

- Runs tasks synchronously
- Executes in same process

---

## ✅ Advantages

- Easy debugging
- Step-by-step execution

---

## ❌ Disadvantages

- No parallelism
- Not for production

---

## 🎯 Use Cases

- Testing
- Debugging task failures

---

# 🚀 Airflow Architecture (Explained from Diagram)

This diagram represents a modern Apache Airflow architecture (Airflow 2.x / 3 style) with distributed execution.

Let’s break it down step-by-step in a clear, interview-ready way 👇

---

# 🧠 1. High-Level Flow

Developer → DAG → DAG Processor → Scheduler → Executor (TRIGGER)→ Queue → Worker → Metadata DB

---

# 🧩 2. Component-by-Component Explanation

---

# 🔹 2.1 Developer (Dev)

Role:
- Writes DAGs (Python files)

Example:
with DAG(...) as dag:
    task1 >> task2

Output:
- DAG files stored in DAG folder

---

# 🔹 2.2 DAGs (Python Files)

Contains:
- Workflow definition
- Tasks
- Dependencies

These are NOT executed directly
They are parsed first

---

# 🔹 2.3 DAG Processor

Very Important

Dedicated process that:
- Reads DAG files
- Parses Python code
- Converts into metadata objects

Why needed?
- Prevents scheduler overload
- Improves performance

---

# 🔹 2.4 Scheduler (Brain)

Role:
- Continuously checks DAGs
- Decides which task should run

What it does:
1. Reads parsed DAGs
2. Creates DAG runs
3. Checks dependencies
4. Sends tasks to execution queue

Does NOT execute tasks

---

# 🔹 2.5 Execution Queue

Temporary holding area

Stores:
- Tasks ready for execution

---

# 🔹 2.6 Executor (Execution Engine)

Role:
- Picks tasks from queue
- Decides how and where to run

Types:
- LocalExecutor
- CeleryExecutor
- KubernetesExecutor

---

# 🔹 2.7 Redis Queue (Celery Setup)

Message broker

Role:
- Stores tasks
- Workers pull tasks

---

# 🔹 2.8 Worker

Actual execution happens here

Responsibilities:
- Executes task code
- Runs scripts
- Updates status

---

# 🔹 2.9 Triggerer

Handles:
- Sensors
- Waiting tasks
- Async execution

Why:
- Frees worker resources

---

# 🔹 2.10 API Layer

Central communication layer

---

# 🔹 2.11 Metadata Database

Stores:
- DAG runs
- Task states
- Logs
- System state

Critical component

---

# 🔄 3. End-to-End Flow

1. Developer creates DAG : Stored in DAG folder
2. DAG Processor parses DAG : Converts into metadata
3. Scheduler schedules tasks : Creates DAG run, Identifies runnable tasks
4. Tasks sent to queue
5. Executor dispatches tasks: Sends to Redis queue (Celery case)
6. Worker executes tasks: Executes task
7. Triggerer handles async tasks: Frees worker resources
8. Metadata DB updated: Stores status
9. UI/API shows results

** ALL THE THINGS ARE DONE VIA API WHICH IS SYNCED WITH AIRLFOW BACKEND WHICH GIVES THE INFO ON UI(FRONTEND)**
---

# ⚡ 4. Key Insights

Separation of roles:
- Scheduler → planning
- Executor → dispatch
- Worker → execution

Queue-based system:
- Enables scalability

Distributed system:
- Supports multiple workers

---

# 🎯 5. Final Summary

Airflow processes DAGs, schedules tasks, distributes execution, and tracks everything in a metadata database.
