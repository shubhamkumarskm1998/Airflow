# ⚙️ 2. Code Examples

---

# 🔥 WITH DAG (Classic)

---

## Daily Cron Example

```python
from airflow import DAG
from datetime import datetime

with DAG(
    dag_id="daily_pipeline",
    start_date=datetime(2024,1,1),
    schedule="0 0 * * *",
    catchup=False
) as dag:
    pass
```

---

## Interval Example

```python
from datetime import timedelta

with DAG(
    dag_id="interval_pipeline",
    start_date=datetime(2024,1,1),
    schedule=timedelta(hours=2),
    catchup=False
) as dag:
    pass
```

---

## Manual Only

```python
with DAG(
    dag_id="manual_pipeline",
    start_date=datetime(2024,1,1),
    schedule=None,
    catchup=False
) as dag:
    pass
```

---

# ⚡ Decorator Style

---

## Daily Example

```python
from airflow.decorators import dag
from datetime import datetime

@dag(start_date=datetime(2024,1,1), schedule="@daily", catchup=False)
def daily_pipeline():
    pass

dag = daily_pipeline()
```

---

## Cron Example

```python
@dag(start_date=datetime(2024,1,1), schedule="*/5 * * * *", catchup=False)
def realtime_pipeline():
    pass

dag = realtime_pipeline()
```

---

## Delta Example

```python
from datetime import timedelta

@dag(start_date=datetime(2024,1,1), schedule=timedelta(hours=2), catchup=False)
def interval_pipeline():
    pass

dag = interval_pipeline()
```

---

# 🎯 3. Best Practices

- Use cron for fixed schedules
- Use timedelta for intervals
- Disable catchup unless needed
- Use event-based for streaming pipelines

# 💥 Final Summary

Airflow supports multiple scheduling strategies for real-world pipelines like batch, streaming, reporting, and manual workflows.
