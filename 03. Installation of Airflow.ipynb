# 🚀 Complete Step-by-Step: Airflow 3.0.3 on EC2 with AWS RDS + Redis

---

# 🧠 1. Architecture

EC2 (Airflow)
   ↓
RDS PostgreSQL (Metadata DB)
   ↓
ElastiCache Redis (Broker)
   ↓
Workers

---

# 🎯 2. Step 1: Launch EC2

- Instance: t3.medium
- OS: Ubuntu
- Open ports:
  - 22 (SSH)
  - 8080 (Airflow UI)

---

# 🔐 3. Step 2: Create RDS PostgreSQL

1. Go to AWS RDS
2. Create database
3. Choose PostgreSQL
4. Set:
   - DB name: airflow
   - Username: airflow_user
   - Password: password
5. Enable public access (or same VPC)

---

# 🔐 4. Step 3: Create Redis (ElastiCache)

1. Go to ElastiCache
2. Create Redis cluster
3. Disable cluster mode
4. Note endpoint

---

# 🔐 5. Step 4: Security Groups

Allow:
- EC2 → RDS (5432)
- EC2 → Redis (6379)
- External → EC2 (8080)

---

# 🔧 6. Step 5: Install Dependencies

sudo apt update
sudo apt install python3-pip python3-venv -y

---

# 📁 7. Step 6: Setup Virtual Environment

mkdir airflow_project
cd airflow_project

python3 -m venv airflow_venv
source airflow_venv/bin/activate

---

# 📦 8. Step 7: Install Airflow

`export AIRFLOW_VERSION=3.0.3
export PYTHON_VERSION=$(python3 --version | cut -d " " -f 2 | cut -d "." -f 1-2)
export CONSTRAINT_URL=https://raw.githubusercontent.com/apache/airflow/constraints-${AIRFLOW_VERSION}/constraints-${PYTHON_VERSION}.txt`

`pip install "apache-airflow==${AIRFLOW_VERSION}" --constraint "${CONSTRAINT_URL}"`

---

# 📂 9. Step 8: Set AIRFLOW_HOME

export AIRFLOW_HOME=~/airflow
mkdir -p $AIRFLOW_HOME

---

# 📦 10. Step 9: Install Extra Packages

pip install psycopg2-binary redis celery

---

# ⚙️ 11. Step 10: Configure airflow.cfg

Update:

executor = CeleryExecutor

sql_alchemy_conn = postgresql+psycopg2://USER:PASSWORD@RDS-ENDPOINT:5432/airflow

broker_url = redis://REDIS-ENDPOINT:6379/0

result_backend = db+postgresql://USER:PASSWORD@RDS-ENDPOINT:5432/airflow

---

# 🗄️ 12. Step 11: Initialize DB

airflow db migrate

---

# 👤 13. Step 12: Create Admin

airflow users create --username admin --password admin --role Admin --email admin@example.com

---

# ▶️ 14. Step 13: Start Services

Terminal 1:
airflow scheduler

Terminal 2:
airflow webserver --port 8080

Terminal 3:
airflow celery worker

---

# 🌐 15. Step 14: Access UI

http://<EC2-IP>:8080

---

# ⚠️ 16. Troubleshooting

- Tasks stuck → worker not running
- DB error → check RDS connection
- Redis error → check SG

---

# 🎯 Final Summary

- EC2 runs Airflow
- RDS stores metadata
- Redis manages task queue
- Workers execute tasks

# 🔧 3. Full airflow.cfg (Important Sections Only)

## CORE SETTINGS

[core]
dags_folder = /home/ubuntu/airflow/dags
executor = CeleryExecutor
parallelism = 32
max_active_tasks_per_dag = 16
max_active_runs_per_dag = 1
load_examples = False

---

## DATABASE

[database]
sql_alchemy_conn = postgresql+psycopg2://USER:PASSWORD@RDS-ENDPOINT:5432/airflow

---

## CELERY

[celery]
broker_url = redis://REDIS-ENDPOINT:6379/0
result_backend = db+postgresql://USER:PASSWORD@RDS-ENDPOINT:5432/airflow
worker_concurrency = 8

---

## WEBSERVER

[webserver]
web_server_port = 8080
web_server_host = 0.0.0.0

---

## SCHEDULER

[scheduler]
scheduler_heartbeat_sec = 5
min_file_process_interval = 30

---

## LOGGING

[logging]
base_log_folder = /home/ubuntu/airflow/logs
remote_logging = True
remote_base_log_folder = s3://your-bucket/airflow-logs

---

# ⚙️ 4. Additional Production Steps

---

## ✅ Enable S3 Logging

pip install apache-airflow-providers-amazon

Set:
remote_logging = True

---

## ✅ Setup IAM Role

- Attach S3 access to EC2
- Avoid access keys

---

## ✅ Use systemd Services

Create service:

sudo nano /etc/systemd/system/airflow-scheduler.service

Example:
[Unit]
Description=Airflow Scheduler

[Service]
User=ubuntu
ExecStart=/home/ubuntu/airflow_project/airflow_venv/bin/airflow scheduler
Restart=always

[Install]
WantedBy=multi-user.target

---

## Enable service

sudo systemctl daemon-reexec
sudo systemctl enable airflow-scheduler
sudo systemctl start airflow-scheduler

---

# 🚀 Install Apache Airflow on Personal PC Using Docker (Best Practice)

---

# 🧠 1. Why Use Docker for Airflow?

- No dependency issues
- Easy setup
- Production-like environment
- Isolation

---

# 🎯 2. Prerequisites

Install:
- Docker
- Docker Compose

---

# 📁 3. Create Project Folder

mkdir airflow-docker
cd airflow-docker

---

# 📥 4. Download Airflow Docker Compose

curl -LfO https://airflow.apache.org/docs/apache-airflow/3.0.3/docker-compose.yaml

---

# 📂 5. Create Folders

mkdir -p ./dags ./logs ./plugins ./config

---

# ⚙️ 6. Create .env File

AIRFLOW_UID=50000

---

# 🔧 7. Initialize Airflow

docker-compose up airflow-init

---

# ▶️ 8. Start Airflow

docker-compose up

---

# 🌐 9. Access UI

http://localhost:8080

Login:
- airflow / airflow

---

# 📁 10. Add DAGs

./dags

---

# 🧪 11. Example DAG

from airflow import DAG
from airflow.operators.python import PythonOperator
from datetime import datetime

def hello():
    print("Hello from Docker Airflow")

with DAG("test_dag", start_date=datetime(2024,1,1), schedule_interval="@daily", catchup=False) as dag:
    task = PythonOperator(task_id="hello", python_callable=hello)

---

# ⚡ 12. Commands

docker-compose up -d
docker-compose logs -f
