# 🧠 1. Event-Based Timetable (Dataset / Event Driven)

---

## 📌 What is Event Timetable?

👉 DAG runs when **event/data is available**, not based on time

---

## 🔥 Example: Dataset Trigger (Airflow 2.4+ / 3.x)

```python
from airflow.decorators import dag, task
from airflow.datasets import Dataset
from pendulum import datetime

dataset = Dataset("s3://data/file.csv")

@dag(
    start_date=datetime(2024,1,1),
    schedule=[dataset],   # event-based trigger
    catchup=False
)
def event_driven_dag():

    @task
    def process():
        print("Triggered by dataset update")

    process()

dag = event_driven_dag()
```

---

## 🧠 How it works

- Another DAG updates dataset
- This DAG triggers automatically

# ⚙️ 2. Production DAG — Incremental (Daily Batch)

---

```python
@dag(
    schedule="0 0 * * *",
    start_date=datetime(2024,1,1),
    catchup=True
)
def prod_incremental():

    @task
    def extract(**kwargs):
        start = kwargs["data_interval_start"]
        end = kwargs["data_interval_end"]
        print(f"Extract {start} → {end}")

    @task
    def load():
        print("Load to warehouse")

    extract() >> load()

dag = prod_incremental()
```

# ⚡ 3. Production DAG — Real-Time (Frequent Cron)

---

```python
@dag(
    schedule="*/5 * * * *",
    start_date=datetime(2024,1,1),
    catchup=False
)
def prod_realtime():

    @task
    def ingest():
        print("Streaming ingestion")

    ingest()

dag = prod_realtime()
```

---

# 🔥 4. Production DAG — Interval (Delta)

---

```python
from datetime import timedelta

@dag(
    schedule=timedelta(hours=2),
    start_date=datetime(2024,1,1),
    catchup=False
)
def prod_interval():

    @task
    def process():
        print("Run every 2 hours")

    process()

dag = prod_interval()
```

---

# 🔹 5. Production DAG — Manual / Adhoc

---

```python
@dag(
    schedule=None,
    start_date=datetime(2024,1,1),
    catchup=False
)
def prod_manual():

    @task
    def run():
        print("Manual execution")

    run()

dag = prod_manual()
```

# 🔄 6. Production DAG — Event Driven Pipeline

---

```python
dataset = Dataset("s3://raw/orders")

@dag(
    schedule=[dataset],
    start_date=datetime(2024,1,1),
    catchup=False
)
def prod_event_pipeline():

    @task
    def transform():
        print("Triggered on new data")

    transform()

dag = prod_event_pipeline()
```

# 🎯 7. When to Use What (Production)

---

| Use Case | Method |
|----------|--------|
| Daily ETL | Cron (@daily) |
| Streaming | Cron (*/5) |
| Fixed interval | Timedelta |
| Event driven | Dataset |
| Debug/adhoc | Manual |

---

# ⚠️ 8. Best Practices

---

- Use incremental logic with data_interval
- Avoid catchup=False for pipelines
- Use dataset for event-driven pipelines
- Keep DAG idempotent

---

# 💥 Final Summary

Airflow supports:
- Time-based (cron, delta)
- Event-based (dataset)
- Manual execution

Each used based on production needs.
